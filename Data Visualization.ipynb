###  ------CLASSIFICATION DATA VISUALIZATION-------

#### Data for classification visualization : Bank Marketing dataset from UCI data repository.

The motive of choosing this data was very similar to the above chosen data.
Banking being a critical application the classification model, this data was chosen so that the features could be more explored and plotted accordingly.<br>
The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed. <br>
The classification goal is to predict if the client will subscribe (yes/no) a term deposit (variable y).<br>
Few of the complex attributes which are not easy to read are stated below : <br>
emp.var.rate: employment variation rate - quarterly indicator (numeric) <br>
cons.price.idx: consumer price index - monthly indicator (numeric)<br>
cons.conf.idx: consumer confidence index - monthly indicator (numeric) <br>
euribor3m: euribor 3 month rate - daily indicator (numeric)<br>
nr.employed: number of employees - quarterly indicator (numeric) <br>


https://archive.ics.uci.edu/ml/datasets/Bank+Marketing#
